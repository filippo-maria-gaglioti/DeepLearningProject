```python
####################### numero totale di incidenti per giorno della settimana #####################

giorni_settimana_ordinati = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Calcola il numero totale di incidenti per giorno della settimana
incidenti_per_giorno_settimana = df['giorno_settimana'].value_counts().reindex(giorni_settimana_ordinati)

# Calcola il numero totale di morti per giorno della settimana
morti_per_giorno_settimana = df.groupby('giorno_settimana')['num_morti'].sum().reindex(giorni_settimana_ordinati)

# Calcola il numero totale di feriti per giorno della settimana
feriti_per_giorno_settimana = df.groupby('giorno_settimana')['num_feriti'].sum().reindex(giorni_settimana_ordinati)

# Visualizza i grafici
fig, axes = plt.subplots(3, 1, figsize=(10, 15))

# Grafico per il numero totale di incidenti per giorno della settimana
axes[0].bar(incidenti_per_giorno_settimana.index, incidenti_per_giorno_settimana, color='skyblue')
axes[0].set_title('Numero totale di incidenti per giorno della settimana')
axes[0].set_xlabel('Giorno della settimana')
axes[0].set_ylabel('Numero di incidenti')
axes[0].grid(axis='y', linestyle='--', alpha=0.7)
axes[0].tick_params(axis='x', rotation=45)

# Grafico per il numero totale di morti per giorno della settimana
axes[1].bar(morti_per_giorno_settimana.index, morti_per_giorno_settimana, color='salmon')
axes[1].set_title('Numero totale di morti per giorno della settimana')
axes[1].set_xlabel('Giorno della settimana')
axes[1].set_ylabel('Numero di morti')
axes[1].grid(axis='y', linestyle='--', alpha=0.7)
axes[1].tick_params(axis='x', rotation=45)

# Grafico per il numero totale di feriti per giorno della settimana
axes[2].bar(feriti_per_giorno_settimana.index, feriti_per_giorno_settimana, color='lightgreen')
axes[2].set_title('Numero totale di feriti per giorno della settimana')
axes[2].set_xlabel('Giorno della settimana')
axes[2].set_ylabel('Numero di feriti')
axes[2].grid(axis='y', linestyle='--', alpha=0.7)
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
#################### GRAFICO GIORNI E MOMENTO GIoRNALIERO NUMERO INCIDENTI #######################

# Definisci l'ordine dei giorni della settimana e dei momenti della giornata
giorni_settimana_ordinati = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
momenti_giornata_ordinati = ['Mattina', 'Pomeriggio', 'Sera', 'Notte']

# Calcola il numero totale di incidenti per giorno della settimana e momento della giornata
incidenti_per_giorno_momento_giornata = df.groupby(['giorno_settimana', 'momento_giornata']).size().unstack(fill_value=0)
incidenti_per_giorno_momento_giornata = incidenti_per_giorno_momento_giornata.reindex(columns=momenti_giornata_ordinati)

# Visualizza il grafico
plt.figure(figsize=(12, 8))
sns.heatmap(incidenti_per_giorno_momento_giornata, cmap='Blues', annot=True, fmt='d', linewidths=.5, cbar=True)
plt.title('Numero totale di incidenti per giorno della settimana e momento della giornata')
plt.xlabel('Momento della giornata')
plt.ylabel('Giorno della settimana')
plt.xticks(ticks=range(len(momenti_giornata_ordinati)), labels=momenti_giornata_ordinati, rotation=45)
plt.yticks(ticks=range(len(giorni_settimana_ordinati)), labels=giorni_settimana_ordinati, rotation=0)
plt.tight_layout()
plt.show()
```

```python
#################### GRAFICO GIORNI E MOMENTO GIoRNALIERO NUMERO MORTI #######################

giorni_settimana_ordinati = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
momenti_giornata_ordinati = ['Mattina', 'Pomeriggio', 'Sera', 'Notte']

# Calcola il numero totale di morti per giorno della settimana e momento della giornata
morti_per_giorno_momento_giornata = df.groupby(['giorno_settimana', 'momento_giornata'])['num_morti'].sum().unstack(fill_value=0)
morti_per_giorno_momento_giornata = morti_per_giorno_momento_giornata.reindex(columns=momenti_giornata_ordinati)

# Visualizza il grafico
plt.figure(figsize=(12, 8))
sns.heatmap(morti_per_giorno_momento_giornata, cmap='Reds', annot=True, fmt='d', linewidths=.5, cbar=True)
plt.title('Numero totale di morti per giorno della settimana e momento della giornata')
plt.xlabel('Momento della giornata')
plt.ylabel('Giorno della settimana')
plt.xticks(ticks=range(len(momenti_giornata_ordinati)), labels=momenti_giornata_ordinati, rotation=45)
plt.yticks(ticks=range(len(giorni_settimana_ordinati)), labels=giorni_settimana_ordinati, rotation=0)
plt.tight_layout()
plt.show()
```

```python
#################### GRAFICO GIORNI E MOMENTO GIoRNALIERO NUMERO FERITI #######################

# Definisci l'ordine dei giorni della settimana e dei momenti della giornata
giorni_settimana_ordinati = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
momenti_giornata_ordinati = ['Mattina', 'Pomeriggio', 'Sera', 'Notte']

# Calcola il numero totale di feriti per giorno della settimana e momento della giornata
feriti_per_giorno_momento_giornata = df.groupby(['giorno_settimana', 'momento_giornata'])['num_feriti'].sum().unstack(fill_value=0)
feriti_per_giorno_momento_giornata = feriti_per_giorno_momento_giornata.reindex(columns=momenti_giornata_ordinati)

# Visualizza il grafico
plt.figure(figsize=(12, 8))
sns.heatmap(feriti_per_giorno_momento_giornata, cmap='Greens', annot=True, fmt='d', linewidths=.5, cbar=True)
plt.title('Numero totale di feriti per giorno della settimana e momento della giornata')
plt.xlabel('Momento della giornata')
plt.ylabel('Giorno della settimana')
plt.xticks(ticks=range(len(momenti_giornata_ordinati)), labels=momenti_giornata_ordinati, rotation=45)
plt.yticks(ticks=range(len(giorni_settimana_ordinati)), labels=giorni_settimana_ordinati, rotation=0)
plt.tight_layout()
plt.show()
```
