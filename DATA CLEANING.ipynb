```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
percorso_file = 'C:\\Users\\pippo\\OneDrive\\Desktop\\Tesi\\incidenti.xlsx'
df = pd.read_excel(percorso_file)
df.columns
```

```python
##### DIVIDERE ORA E DATA INCIDENTE E FARE DUE COLONNE SEPARATE ########

df['dataoraincidente'] = df['dataoraincidente'].astype(str)
df[['data_incidente', 'ora_incidente']] = df['dataoraincidente'].str.split(' ', 1, expand=True)

colonne_da_spostare=df.columns[-2:]
for colonna in colonne_da_spostare[::-1]:
    df.insert(0, colonna, df.pop(colonna))
    
#ELIMINARE COLONNE SUPERFLUE############    
df = df.drop('idprotocollo', axis=1)
df= df.drop('dataoraincidente', axis=1)

df.head(100)
```

```python
####### AGGIUNTA COLONNA GIORNO DELLA SETTIMANA ###############

def giorno_settimana_da_data(data_stringa):
    data = datetime.strptime(data_stringa, '%Y-%m-%d')
    return data.strftime('%A')

df['giorno_settimana'] = df['data_incidente'].apply(giorno_settimana_da_data)
indice_colonna_data = df.columns.get_loc('data_incidente')
df.insert(indice_colonna_data + 1, 'giorno_settimana', df.pop('giorno_settimana'))
df.head(100)
```

```python
##########  CATEGORIZZAZIONE ORARIO SERA; MATTIN; NOTTE #############


def categorize_momento_giornata(ora):
    ora = int(ora.split(':')[0])
    if 6 <= ora < 14:
        return 'Mattina'
    elif 14 <= ora < 18:
        return 'Pomeriggio'
    elif 18 <= ora < 23:
        return 'Sera'
    else:
        return 'Notte'

df['momento_giornata'] = df['ora_incidente'].apply(categorize_momento_giornata)

indice_colonna_ora_incidente = df.columns.get_loc('ora_incidente')
df.insert(indice_colonna_ora_incidente + 1, 'momento_giornata', df.pop('momento_giornata'))

df.head(100)
```

```python
############### SOSTITUISCI TUTTE LE OCCORRENZE DI "Più di due carreggiate" CON "Più di due carreggiate" ######################
df['tipostrada'] = df['tipostrada'].replace('PiÃ¹ di due carreggiate', 'Più di due carreggiate')
```

```python
######################### INTRODUZIONE FEATURE INCROCIO ##############################################################
# Definisci una funzione per determinare il valore della colonna "incrocio"
def determina_incrocio(row):
    if pd.notnull(row['strada1']) and pd.notnull(row['strada2']):
        return 1
    elif pd.notnull(row['strada1']) and pd.isnull(row['strada2']):
        return 0
    else:
        return None

# Applica la funzione alla riga del DataFrame e crea la nuova colonna "incrocio"
df['incrocio'] = df.apply(determina_incrocio, axis=1)
```
